Loop through the Data, extract accuracy and reaction time

```python
import pandas as pd

df = pd.read_excel("AkshayEvents.xlsx", header=None)

data = []
time = None
accuracy = None

for line in df[0]:

    line = str(line).strip()

    if "#OnlineScore" in line:


        if "Correct" in line:
            accuracy = "Correct"
        elif "Wrong" in line:
            accuracy = "Wrong"
        else:
            accuracy = None


        if "Time -" in line:
            parts = line.split("Time -")
            if len(parts) > 1:
                time = parts[1].split(",")[0].strip()

    elif line.startswith("F") and time:
        data.append([line, time, accuracy])
        time = None
        accuracy = None


df_clean = pd.DataFrame(data, columns=["Trial", "ResponseTime", "Accuracy"])

print("Rows extracted:", len(df_clean))
df_clean.head()
```
